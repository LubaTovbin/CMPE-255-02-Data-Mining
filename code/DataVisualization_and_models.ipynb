#### Getting features and label and performing few transformations

```python
X = reader.get_features('dropout', 'deviation')
y = reader.get_label('total')
y = np.power(np.log1p(y),1/2)
```

#### Splitting the dataset into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Applying various regression models on the training and test set

#### Linear Regression

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred=linreg.predict(X_test)
```

```python
linmse=metrics.mean_squared_error(y_test, y_pred)
linmse
```

```python
mae=metrics.mean_absolute_error(y_test, y_pred)
mae
```

```python
linscore=linreg.score(X_test,y_test)
linscore
```

#### Ridge Regression

```python
ridgereg = Ridge(alpha=1.0)
ridgereg=ridgereg.fit(X_train,y_train)
y_pred=ridgereg.predict(X_test)
```

```python
ridmse=metrics.mean_squared_error(y_test, y_pred)
ridmse
```

```python
mae=metrics.mean_absolute_error(y_test, y_pred)
mae
```

```python
ridscore=ridgereg.score(X_test,y_test)
ridscore
```

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth=2)
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
```

```python
dtmse=metrics.mean_squared_error(y_test, y_pred)
dtmse
```

```python
mae=metrics.mean_absolute_error(y_test, y_pred)
mae
```

```python
dtscore=regr.score(X_test,y_test)
dtscore
```

#### K nearest neighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X, y) 
y_pred=neigh.predict(X_test)
```

```python
knmse=metrics.mean_squared_error(y_test, y_pred)
knmse
```

```python
mae=metrics.mean_absolute_error(y_test, y_pred)
mae
```

```python
knscore=neigh.score(X_test,y_test)
knscore
```

#### Support Vector Regression

```python
from sklearn.svm import SVR
svr = SVR(gamma='auto')
svr=svr.fit(X_train, y_train.values.ravel())
y_pred=svr.predict(X_test)
```

```python
svmse=metrics.mean_squared_error(y_test, y_pred)
svmse
```

```python
mae=metrics.mean_absolute_error(y_test, y_pred)
mae
```

```python
svrscore=svr.score(X_train,y_train)
svrscore
```

### Comparing accuracy of all the above models

```python
from tabulate import tabulate
table = [["Linear Regression",linscore,linmse],["Ridge Regression",ridscore,ridmse],["Decision Tree Regression",dtscore,dtmse],["K-Nearest Neighbours Regression",knscore,knmse],["Support Vector Regression",svrscore,svmse]]
print(tabulate(table, headers=["Regression Model","Accuracy Score (%)", "Mean Squared Error"]))
    
```

### Support Vector Regression model has the highest accuracy score (92.92 %) amongst all the models tried.
