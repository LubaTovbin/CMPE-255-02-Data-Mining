```python
Stdev_list = []
for i in income_df.index:
    Stdev_list.append(float(income_df.at[0,'Stdev'][:-3]))
```

```python
income_df['Stdev_not_sqrd'] = pd.Series(Stdev_list)
income_df['Stdev_sqrd'] = income_df['Stdev_not_sqrd']**2
```
