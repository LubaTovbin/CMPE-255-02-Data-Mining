```python
#build a dictionary of separate income dataframes for each state
income_df_dict = {}
for i in crime_df.index:
    st = crime_df.at[i,'State_ab']
    state_income_df = income_df.loc[income_df['State_ab'] == st]
    income_df_dict.update({st:state_income_df}) 
```

```python
mean_list = []
median_list = []
stdv_list = []
households_list = []

for i in crime_df.index:
    st = crime_df.at[i,'State_ab']
    plce = crime_df.at[i,'Place']
    
    for j in income_df_dict[st].index:              
        if plce==income_df_dict[st].at[j, 'County']:    #or plce==income_df_dict[st].at[j, 'Place']:
            mean_list.append(income_df_dict[st].at[j,'Mean'])
            median_list.append(income_df_dict[st].at[j,'Median'])
            stdv_list.append(income_df_dict[st].at[j,'Stdev'])
            households_list.append(income_df_dict[st].at[j,'Households'])
```

```python
print(len(mean_list))
print(len(median_list))
print(len(stdv_list))
print(len(households_list))
```
